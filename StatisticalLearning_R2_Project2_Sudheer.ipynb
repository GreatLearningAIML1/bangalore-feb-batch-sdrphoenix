```python
import numpy as np
import pandas as pd
from scipy.stats import t, ttest_1samp, ttest_ind, mannwhitneyu, levene, shapiro, wilcoxon, ttest_rel
from statsmodels.stats.power import ttest_power
```

```python
data = pd.read_csv('data.csv')
```

```python
#Answer for Q1
print("Mean of Old scheme (in thousands) - ", data["Old Scheme (in thousands)"].mean())
print("Mean of New scheme (in thousands) - ", data["New Scheme (in thousands)"].mean())
```

```python
# one-tail Paired t-test is required as we are checking whether output has been significantly "Raised" in new scheme
group1 = data["Old Scheme (in thousands)"]
group2 = data["New Scheme (in thousands)"]
t_statistic, p_value = ttest_rel(group1,group2)

# Divide p_value by 2 to get significance value for one-tailed (as we are testing only for raise of output sales)
p_value = p_value / 2

# Answer for Q2
print(" Answer for Q2 - p_value for one-tailed ttest =", p_value)
```

```python
#Answer for Q3
print("Answer for Q3 - Conclusion: p_value 0.06528776980668831 > 0.05")
print("Null Hypothesis is accepted i.e., There is NO significant difference in the two schemes w.r.t output sales ")
```

```python
#Answer for Q-4a 
#Probability of type-1 error
print("Probability of Type-1 error is alpha i.e., 0.05")
```

```python
#Answers for Question 4b & 4c
# Determine t_critical value using:-
# confidence coefficient 1-alpha i.e., 1-0.05 = 0.95
# degrees of freedom (df) = n-1 = 30-1 =29

conf = 0.95
df = 29
t_critical = t.ppf(conf,df)
print("t_critical=",t_critical)

# Determine Xbar using test-statistic formula i.e, test-stat = (Xbar-mu0)/(S/sqrt(n))
Xbar = t_critical * np.std(group2-group1,ddof=1)/np.sqrt(30)

#Determine test-statistic value with mean difference of 5 thousand
test_stat = (Xbar-5.0)/(np.std(group2-group1,ddof=1)/np.sqrt(30))
print("test_stat =", test_stat)

#Determine p-value - Answer for Question-4b
p_val = (1.0 - t.cdf(abs(test_stat), df))
print("Answer for Question 4b - p-value =", p_val)

# #Determine Power - Answer for Question-4c 
power = 1 - t.cdf(test_stat, df) 
print("Answer for Question 4c - Power of Test =", power)
```
