## Question 1 

*A student is trying to decide between two GPUs. He want to use the GPU for his research to run Deep learning algorithms, so the only thing he is concerned with is speed.*

*He picks a Deep Learning algorithm on a large data set and runs it on both GPUs 15 times, timing each run in hours. Results are given in the below lists GPU1 and GPU2.*

Hint: You can import ttest function from scipy to perform t tests 

Given,

Null Hypothesis : There is no significant difference between data sets

Alternate Hypothesis : There is a significant difference

*Do two-sample testing and check whether to reject Null Hypothesis or not.*

#### Answer:

```python
import pandas as pd
import scipy.stats as stats
from scipy.stats import ttest_1samp, ttest_ind, levene, shapiro, f_oneway
import matplotlib.pyplot as plt
GPU1 = pd.DataFrame([11,9,10,11,10,12,9,11,12,9,11,12,9,10,9])
GPU2 = pd.DataFrame([11,13,10,13,12,9,11,12,12,11,12,12,10,11,13])
```

```python
t_statistic, p_value = ttest_ind(GPU1, GPU2)
print(t_statistic,p_value)
#Reject Null-Hypothesis as p_value = 0.01379428 is less than 0.05
#Accept Alternate Hypothesis i.e., there is significant difference between GPU1 and GPU2
```

If you need to compare more than two data sets at a time, an ANOVA is your best bet. 

*The results from three experiments with overlapping 95% confidence intervals are given below, and we want to confirm that the results for all three experiments are not significantly different.*

#Assumption: All the 3 datasets (e1,e2 & e3) are random, independent, parametric & normally distributed
But before conducting ANOVA, test equality of variances (using Levene's test) is satisfied or not. If not, then mention that we cannot depend on the result of ANOVA

```python
import pandas as pd
e1 = pd.DataFrame([1.595440,1.419730,0.000000,0.000000])
e2 = pd.DataFrame([1.433800,2.079700,0.892139,2.384740])
e3 = pd.DataFrame([0.036930,0.938018,0.995956,1.006970])
```

Hint - You can use stats.levene function and stats.f_oneway function.

#### Answer:

```python
levene(e1,e2,e3)
#Using Levene's test equality of variances is satisfied as pvalue (0.12259793) > 0.05
```

```python
f, p = f_oneway(e1,e2,e3)

print ('One-way ANOVA')
print ('=============')
 
print ('F value:', f)
print ('P value:', p, '\n')

# Since p-value (0.13574645) > 0.05 , accept Null Hypothesis 
# Thereby, It can be confirmed that all three experiments are not significantly different
```

# Question 12 - Find coefficients & intercept
Estimate the coefficients b0 and b1 using scikit-learn.

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(xTrain, yTrain)
```

```python
b0 = lr.coef_
b1 = lr.intercept_
print('b0 Coefficient =', b0)
print('b1 Intercept =', b1)
```

Check: coef_ and intercept_ functions can help you get coefficients & intercept

# Question 13 - Linear Relationship between feature and target 
Plot the line with b1 and b0 as slope and y-intercept.

Hint - y = mx + c, plot y

```python
import matplotlib.pyplot as plt

def func(x):
    return b0 * x + b1
line = func(xTrain)
plt.scatter(xTrain, yTrain)
plt.plot(xTrain, line, c= 'r')
plt.show()
```

# Question 14 - Evaluation of model with scikit-learn 
Validate the model with Root Mean Squares error and R^2 score using scikit-learn. RMSE and R2 for test data and prediction

Hint: You can import mean_squared_error function & r2 (R square) from sklearn.metrics. Performing root operation over mean square error over mean square error gives you root mean square error

```python
y_pred = lr.predict(xTest)

from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt

mse = mean_squared_error(yTest, y_pred)
rmse = sqrt(mse)

print('Root Mean Squares Error =' , rmse)
```

```python
rsqr = r2_score(yTest, y_pred)
print('R-Square = ' , rsqr)
```
